```python
import warnings
import seaborn as sns
import sys
import os

sys.path.append(os.path.abspath('../'))

from dynamo.plot_shannon_entropy import DiversityAnalysis

warnings.filterwarnings('ignore')
sns.set_style('white')
```

```python
subjects = ['male', 'female', 'donorA', 'donorB']
```

```python
wd =  './data/data_original/alpha_diversity/shannon/'
```

```python
# Shannon Diversity Index
shannon_file_paths = {
   'male': wd + 'male_shannon_entropy.csv',
   'female': wd + 'female_shannon_entropy.csv',
   'donorA': wd + 'donorA_shannon_entropy.csv',
   'donorB': wd + 'donorB_shannon_entropy.csv'
}
```

```python
shannon_analyzer = DiversityAnalysis(shannon_file_paths, subjects)
```

```python

breakpoints = {
    'female': [60],
    'donorB': [150]
}
highlights = {
    'donorA': [(70, 122)],
    'donorB': [(150, 159)]
}
```

```python
shannon_analyzer.plot_trends(breakpoints, highlights, title="Shannon diversity index")
```
